#Interview file

Here is a description of the columns in the provided dataset:

AID: every entry stores a unique whole number assigned to a job applicant for the purpose of identification
Name: every entry stores the first of an applicant
English: every entry contains the marks obtained out of 100 in the English test by an applicant
Mathematics: every entry contains the marks obtained out of 100 in the mathematics test by an applicant
Programming: every entry contains the marks obtained out of 100 in the programming test by an applicant

```python
df = pd.read_csv('/content/Interview_First_Stage_Exam.csv',index_col=0)
df.head()
```

```python
df.English.median()
```

```python
df.Mathematics.quantile(0.75)
```

```python
# prompt: Which of the following candidates were eliminated after the first round? Note that the first quartile of each subject is the threshold value to qualify for the second or third round of interview. (Note: More than one option may be correct.)

english_threshold = df['English'].quantile(0.25)
mathematics_threshold = df['Mathematics'].quantile(0.25)
programming_threshold = df['Programming'].quantile(0.25)

eliminated_candidates = df[(df['English'] < english_threshold) |
                           (df['Mathematics'] < mathematics_threshold) |
                           (df['Programming'] < programming_threshold)]

print("Candidates eliminated after the first round:")
print(eliminated_candidates[['Name']])
```

```python
# prompt: The upper cut-off marks in mathematics, English and programming are Equation, Equation and Equation, respectively. Candidates who scored more than the upper cut-off marks will directly go to the third round. According to this decision, how many candidates are exempted from the second round of the interview and will proceed directly to the third round?

english_upper_cutoff = df['English'].quantile(0.75)
mathematics_upper_cutoff = df['Mathematics'].quantile(0.75)
programming_upper_cutoff = df['Programming'].quantile(0.75)

exempted_candidates = df[(df['English'] > 76.5) &
                         (df['Mathematics'] > 77.25) &
                         (df['Programming'] > 75)]

print(f"\nNumber of candidates exempted from the second round: {len(exempted_candidates)}")
```

Investment Returns

Here is a description of the columns in the provided dataset:

- Year: every entry stores the year for which an investment return was calculated
- George and Wallace: every entry stores the percentage return for an investment via the company "George and Wallace"
- Imperial Traders: every entry stores the percentage return for an investment via the company "Imperial Traders"
- High Rise: every entry stores the percentage return for an investment via the company "High Rise"

```python
df = pd.read_csv('/content/Investment_Returns.csv')
df.head()
```

```python
df['High Rise'].min()
```

```python
df['High Rise'].max()
```

```python
# prompt: What is the range of the percentage returns by the company High Rise? (Note: Round your answer to the nearest whole number.)

high_rise_range = df['High Rise'].max() - df['High Rise'].min()
print(f"The range of percentage returns for High Rise is: {round(high_rise_range)}")
```

```python
# prompt: You would have probably noticed that in the Jupyter Notebook, the mean absolute deviation about the mean is calculated by default. According to the above-mentioned statement, what is the mean absolute deviation about the mean of the percentage returns by the company Imperial Traders? (Note: Round your answer to two decimal places.)

# Calculate the mean absolute deviation about the mean for 'Imperial Traders'
imperial_traders_returns = df['Imperial Traders']
mean_imperial_traders = imperial_traders_returns.mean()

mad_imperial_traders = (imperial_traders_returns - mean_imperial_traders).abs().mean()

print(f"The mean absolute deviation about the mean for Imperial Traders is: {round(mad_imperial_traders, 2)}")
```

```python
ITR = df['Imperial Traders']
mean_ITR = ITR.mean()
mad_ITR = (ITR - mean_ITR).abs().mean()
mad_ITR
```
